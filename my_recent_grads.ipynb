##### Q. Is there any link between the number of full-time employees and median salary?

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Median salary full-time  vs. Sample_size')
```

Weak correlation between 0-1000 sample_size as well as median salary 

```python
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Unemployment Rate vs. Sample_size')
```

```python
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Full-time  vs. Median')
#ax3.set_xlim(0,50000)
```

Weak negative correlation between 0-50000 full time employees as well as median salary .

```python
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_title('ShareWomen  vs. Unemployment Rate')
#ax3.set_xlim(0,50000)
```

Weak positive correlation 

```python
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax5.set_title('Men vs Median')
#ax5.set_ylim(20000,80000)
#ax5.set_xlim(0,10000)
```

```python
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax6.set_title('Women vs Median')
#ax6.set_ylim(20000,80000)
#ax6.set_xlim(0,10000)
```

setting xlim to  5000
there is stronger correlation between men vs median as compared to women vs median
